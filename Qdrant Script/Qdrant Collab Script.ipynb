```python
import os

from google.colab import userdata, files

from qdrant_client import QdrantClient, AsyncQdrantClient
from qdrant_client.http.models import Distance, VectorParams, PointStruct, UpdateStatus, CollectionStatus

from typing import List
import uuid

from transformers import AutoTokenizer, AutoModel, AutoModelForCausalLM
import torch

from qdrant_client.models import PointStruct

from datasets import load_dataset, Dataset
import pandas as pd

DATA_DIR = "data"
COLLECTION_NAME = "vh_collection"
MODEL_ID = "all-MiniLM-L6-v2"
VECTOR_SIZE = 384  # Change according to embedding model

import os

os.environ["QDRANT_URL"] = userdata.get("QDRANT_URL")
os.environ["QDRANT_API_KEY"] = userdata.get("QDRANT_API_KEY")

import httpx
try:
    response = httpx.get(os.getenv("QDRANT_URL"))
    print(f"Qdrant Server Status: {response.status_code}")
except Exception as e:
    print(f"Error connecting to Qdrant server: {e}")

# Ensure data directory exists
os.makedirs(DATA_DIR, exist_ok=True)

# Upload files (in Google Colab)
from google.colab import files
uploaded_files = files.upload()

# Save uploaded files to data directory
for filename in uploaded_files.keys():
    os.rename(filename, os.path.join(DATA_DIR, filename))

# Load text files
txt_files = [f for f in os.listdir(DATA_DIR) if f.endswith('.txt')]
data = {"title": [], "source": [], "text": []}

for file in txt_files:
    file_path = os.path.join(DATA_DIR, file)
    with open(file_path, "r", encoding="utf-8") as f:
        content = f.read()
        data["title"].append(file)
        data["source"].append(file_path)
        data["text"].append(content)

# Create a Pandas DataFrame and convert to Hugging Face Dataset
df = pd.DataFrame(data)
dataset = Dataset.from_pandas(df)

# Initialize embedding model
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer(MODEL_ID)

print(embedding_model.get_sentence_embedding_dimension())

def get_text_embedding(text: str) -> List[float]:
    """Generate embedding for the input text."""
    return embedding_model.encode(text).tolist()

# Initialize Qdrant client
qdrant_client = QdrantClient(
    url=os.getenv("QDRANT_URL"),
    api_key=os.getenv("QDRANT_API_KEY")
)

Create collection in Qdrant
if not qdrant_client.collection_exists(collection_name=COLLECTION_NAME):
    qdrant_client.create_collection(
        collection_name=COLLECTION_NAME,
        vectors_config=VectorParams(size=VECTOR_SIZE, distance=Distance.COSINE),
    )
else:
    print(f"Collection '{COLLECTION_NAME}' already exists.")

## Use to recreate collection if necessary
# if qdrant_client.collection_exists(collection_name=COLLECTION_NAME):
#     qdrant_client.delete_collection(collection_name=COLLECTION_NAME)

# qdrant_client.create_collection(
#     collection_name=COLLECTION_NAME,
#     vectors_config=VectorParams(size=VECTOR_SIZE, distance=Distance.COSINE),
# )

def add_data_to_qdrant(dataset: Dataset, client: QdrantClient, collection_name: str):
    """Insert data into Qdrant."""
    points = []
    for idx, row in enumerate(dataset):
        text_id = str(uuid.uuid4())
        text = row["text"]
        embedding = get_text_embedding(text)

        # Create payload
        payload = {
            "id": text_id,
            "title": row["title"],
            "source": row["source"],
            "text": text,
        }

        # Prepare point for insertion
        point = PointStruct(
            id=text_id,
            vector=embedding,
            payload=payload,
        )
        points.append(point)

    # Insert points into Qdrant
    operation_info = client.upsert(
        collection_name=collection_name,
        points=points,
        wait=True,
    )

    # Validate insertion
    if operation_info.status == UpdateStatus.COMPLETED:
        print(f"Successfully inserted {len(points)} points.")
    else:
        print("Data insertion failed.")

# Insert data into Qdrant
add_data_to_qdrant(dataset, qdrant_client, COLLECTION_NAME)

# Close Qdrant client
qdrant_client.close()
```
